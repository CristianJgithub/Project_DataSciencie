```python
import pandas as pd
```

```python
df = pd.read_csv('empleados.csv')
```

```python
df.head(5)
```

```python
df.drop('Apellido', axis=1, inplace=True)  # Aqui eliminamos la columna Apellidos del archovo original para el desafio.
```

```python
df.head(5)
```

```python
df[df.isnull().any(axis=1)] # Aqui podemos verificar cuantas filas y columnas contienen al menos un valor faltante. 'nulo'
```

```python
df_actualizado = df.dropna(axis=0) # Aquí eliminamos esas filas que contienen valores nulos.
```

```python
df_actualizado
```

```python
df_actualizado = df.dropna(axis=0, thresh=5)  # Aqui eliminamos los valores nulos pero nos aseguramos de que almenos cada fila Contengan por lo menos cinco valores validos.
```
